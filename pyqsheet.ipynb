### Topic 1: Sets, Relations, and Functions
**A1:** The formula for the power set is $2^n$. Therefore, $2^4 = \mathbf{16}$.
**A2:** Substitute $g(x)$ directly into $f(x)$: $f(x^2) = 3(x^2) - 4 = \mathbf{3x^2 - 4}$.
**A3:** Using the quadratic formula $x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$: $x = \frac{6 \pm \sqrt{36 - 16}}{2} = \mathbf{3 \pm \sqrt{5}}$.
**A4:** For $g(y)$ to be defined, $y^2 - 9 \ge 0 \Rightarrow y \in (-\infty, -3] \cup [3, \infty)$. For $f(g(x))$ to be defined, the denominator $2g(x) \neq 0$, so $g(x) \neq 0$. This means $y \neq 3$ and $y \neq -3$. The final valid domain is **$(-\infty, -3) \cup (3, \infty)$**.
**A5:** Total = 500. Knowing zero languages = 50. Total inside the sets = 450.
Using the Inclusion-Exclusion Principle: $n(P \cup C \cup J) = n(P) + n(C) + n(J) - n(P \cap C) - n(C \cap J) - n(P \cap J) + n(P \cap C \cap J)$. 
$450 = 285 + 195 + 115 - 70 - 50 - 45 + n(P \cap C \cap J)$.
$450 = 595 - 165 + n(P \cap C \cap J) \Rightarrow n(P \cap C \cap J) = 20$.
To find exactly one language, subtract the overlaps from the totals:
* Only Python = $285 - 70 - 45 + 20 = 190$
* Only C = $195 - 70 - 50 + 20 = 95$
* Only Java = $115 - 45 - 50 + 20 = 40$
Sum = $190 + 95 + 40 =$ **325 candidates**.

### Topic 2: Coordinate Geometry & Straight Lines
**A6:** Distance $d = \sqrt{(-1 - 3)^2 + (5 - (-2))^2} = \sqrt{(-4)^2 + (7)^2} = \sqrt{16 + 49} = \mathbf{\sqrt{65}}$.
**A7:** Ajay's pattern: (3 East, 9 North) repeated 3 times. Ajay's final location is $3 \times (3, 9) = (9, 27)$. Ankit is at $(3, 3)$. 
The slope $m$ between them is $\frac{27 - 3}{9 - 3} = \frac{24}{6} = 4$. 
Equation of the line: $y - 3 = 4(x - 3) \Rightarrow \mathbf{y = 4x - 9}$.
**A8:** $\tan \theta = |\frac{m_2 - m_1}{1 + m_1m_2}|$. Slope of $L_1$ is $\frac{6-2}{3-1} = \frac{4}{2} = 2$. For perpendicular lines, $m_1m_2 = -1$. Therefore, the slope of $L_2$ is **$-\frac{1}{2}$**.

**A10:** * Part A: Vertex $x$-coordinate is $-\frac{b}{2a} = -\frac{4}{2(1)} = -2$. The $y$-coordinate is $(-2)^2 + 4(-2) + 12 = 4 - 8 + 12 = 8$. The vertex is at **$(-2, 8)$**.

### Topic 4: Graph Theory
**A13:** Using the Handshaking Lemma: $\sum \text{degrees} = 2 \times \text{edges}$. 
If there are $v$ vertices each with degree 2, then $2v = 2 \times 8 \Rightarrow 2v = 16 \Rightarrow \mathbf{v = 8}$ **vertices**.
**A14:** A minimum vertex cover is the smallest set of vertices such that every single edge in the graph is connected to at least one vertex in the set. For a 4-vertex square graph, selecting any **2** diagonal vertices successfully covers all 4 edges.
**A15:** * Part A: A spanning tree for any graph with $n$ vertices will strictly contain **$n - 1$ edges**. 
* Part B: By the fundamental definition of a tree structure in graph theory, it will contain **zero cycles**.
